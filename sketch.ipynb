#### There are basically 4 steps:
 1. Convert the RGB color image to grayscale.
 2. Invert the grayscale image to get a negative.
 3. Apply a blur to the negative from step 2.
 4. Blend the grayscale image with the blurred negative using a color dodge.



```python
import imageio as im
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#STEP1
img = cv2.imread("meet.jpg")
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# img_gray=cv2.resize(img_gray, (0,0), fx=0.5, fy=0.5)
#the commented code can be enabled if the image is very big
plt.imshow(img_gray)
```

```python
#STEP2
img_gray_inv = 255 - img_gray
plt.imshow(img_gray_inv)
```

```python
#STEP3
img_blur = cv2.blur(img_gray_inv,(40,40))
plt.imshow(img_blur)
```

```python
#Step4
def dodge(image, mask):
    return cv2.divide(image, 255-mask, scale=256)
final_img=dodge(img_gray,img_blur)
plt.imshow(final_img, cmap="gray")
```

```python
plt.imsave('meet_sketch.jpg', final_img, cmap='gray')
```
